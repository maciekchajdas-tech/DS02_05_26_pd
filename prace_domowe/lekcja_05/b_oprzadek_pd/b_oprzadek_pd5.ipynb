# Zadanie 9  Confusion matrix i metryki
Masz confusion matrix dla klasyfikacji binarnej:

cm = [
[80, 10],   # TN, FP

[5, 45]     # FN, TP
]

Wyciągnij wartości TN, FP, FN, TP przez dwuwymiarowe indeksowanie i oblicz:
* accuracy = (TP + TN) / (TP + TN + FP + FN),
* precision = TP / (TP + FP),
* recall = TP / (TP + FN)

Wyświetl wyniki w formacie procentowym z dwoma miejscami po przecinku.

```python
cm = [ [80, 10], # TN, FP
        [5, 45] # FN, TP
      ]
TN = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TP = cm[1][1]
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
print(f"Accuracy : {round(accuracy*100,2)}%, Precision : {round(precision*100,2)}%, Recall : {round(recall*100,2)}%")
```
